'2.1. Теоретический материал – Условный оператор'
#If – Условный оператор
#В коде часто приходится проверять выполнимость или
#невыполнимость каких-то условий. Синтаксис:
#if <условие1 (булевское выражение)> :
# - <код, который выполнится, если условие верно>
#elif <условие2 (булевское выражение)>:
# - <код, который выполнится, если условие1 было неверно, а
#условие2 верно>
#else:
# - <код, который выполнится, если условие1 и условие2 были
#неверны>
#Обратите внимание, что код, который должен выполняться внутри
#каждого условия, записывается с отступом в 4 пробела от уровня if, elif и
#else: в питоне области видимости переменных обозначаются отступами.
#То есть, отступы позволяют понять, где начинается код,
#который должен выполняться при выполнении условия в if, и где
#заканчивается.

```python
'3.1. Теоретический материал – Циклы, массивы, срезы'
#Циклы - for и while
#Цикл while проверяет истинность условия, и если условие истинно, то
#выполняет инструкции цикла.
#    Синтаксис:
#        while условие:
#        - инструкции
#    Инструкции, которые относятся к циклу while, располагаются на
#последующих строках и должны иметь отступ от начала ключевого слова
#while.
#    Цикл for проходит по набору значений, помещает каждое значение в
#переменную с которой в цикле можно производить различные действия.
#    Синтаксис:
#    for переменная in набор_значений:
#    - инструкции
#    После ключевого слова for идет название переменной, в которую
#будут помещаться значения. После оператора in указывается набор значений
#и двоеточие. Инструкции, которые относятся к циклу, располагаются на
#последующих строках и должны иметь отступ.
#    При создании цикла удобно пользоваться функцией range(a, b),
#которая создает последовательность чисел от a до b-1.
```

```python
'4.3.2 Задание. Дана функция Создать массив из 10 значений функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой половины массива и построить графики для основного массива – линейный и для среза – точечный'
import math as m
import numpy as np
import matplotlib.pyplot as plt
def bigfunc(x):
    f = m.sqrt(1 + m.exp(m.sqrt(x)) + m.cos(m.pow(x, 2)))/m.fabs(1 - m.pow(m.sin(x), 3)) + m.log(m.fabs(2*x))
    f = float(f)
    return f
masd = []
countm = []
counts = []
for i in range (1, 11):
    masd.append(bigfunc(i))
    countm.append(i)
    if i < 6:
        counts.append(i)
srez = []
for i in range(0, 5):
    srez.append(masd[i])
print(list(masd), '|', list(srez))
print(list(countm))
print(list(counts))
plt.plot(countm, masd) 
plt.show() 
plt.grid() 
plt.scatter(counts, srez)
```

```python
'4.3.3 Задание. Дана функция |cos(𝑥𝑒(cos(𝑥)+ln(𝑥+1)))| Построить график наинтервале (0,10) с шагом 1 с заливкой площади и найти эту площадь '
import math as m
from scipy.integrate import simps
from numpy import trapz
import matplotlib.pyplot as plt
def bigfunc(x):
    f = m.fabs(m.cos(x * m.exp(m.cos(x) + m.log(x + 1))))
    f = float(f)
    return f
masd = []
count = []
for i in range(0, 11):
    masd.append(bigfunc(i))
    count.append(i)
plt.plot(count, masd)
plt.fill_between(count, masd)
area = trapz(masd)
print(area)
```

```python
'4.3.5 Задание. Создайте простейший калькулятор, включающий основные действиядля двух переменных ′ + ′, ′ − ′, ′ ∙ ′, ′ ÷ ′, а также вычислениеследующих функций:'
import math as m
print(" 0 - завершить работу программы | 1 - вычислить функцию e^(x+y) | 2 - вычислить функцию sin(x+y) | 3 - вычислить функцию cos(x+y) | 4 - вычислить функцию x^y")
def efunc(x, y):
    f = m.exp(x + y)
    f = float(f)
    return f
def sinfunc(x, y):
    f = m.sin(x + y)
    f = float(f)
    return f
def cosfunc(x, y):
    f = m.cos(x + y)
    f = float(f)
    return f
def powfunc(x, y):
    f = m.pow(x, y)
    f = float(f)
    return f
while True:
    key = input("Введите (+,-,*,/): ")
    if key == '0':
        break
    if key in ('+', '-', '*', '/'):
        x = float(input("x="))
        y = float(input("y="))
        if key == '+':
            print("%.2f" % (x+y))
        elif key == '-':
            print("%.2f" % (x-y))
        elif key == '*':
            print("%.2f" % (x*y))
        elif key == '/':
            if y != 0:
                print("%.2f" % (x/y))
            else:
                print("Деление на ноль!")
    if key == '1':
        x = float(input("x="))
        y = float(input("y="))
        print("%.2f" % (efunc(x, y)))
    if key == '2':
        x = float(input("x="))
        y = float(input("y="))
        print("%.2f" % (sinfunc(x, y)))
    if key == '3':
        x = float(input("x="))
        y = float(input("y="))
        print("%.2f" % (cosfunc(x, y)))
    if key == '4':
        x = float(input("x="))
        y = float(input("y="))
        print("%.2f" % (powfunc(x, y)))
    else:
        print("///")
```
